### Research Question
Can we use the medical data set to identify which variables explain the variation in the variable 'Additional_charges' charged to the patient for their care?

### Goals of Analysis

#### Organizational Goals: 

The primary stakeholders, hospital administrators and owners, would be the primary beneficiaries of this kind of analysis. Knowing which patient factors both positively and negatively contribute to Additional Charges could help shape patient care plans and help to identify ways which costs can be managed. For example, if a medical condition is identified as a leading factor in increaing charges, hospital administration can use this information to identify ways in which the condition can be managed in the hospital without contributing to additional charges. An example might be to keep patients with that condition in the same ward to reduce staffing constraints of managing their specialty care. Reducing Additional Charges helps control costs to the hospital and charges to the patient. 


#### Logistical Goals: 

The goals of this analysis are to perform a linear regression analysis on the medical dataset provided. This will begin by first cleaning the data by looking for duplicates, missing values and outliers. A close study will be made of all the considered variables, reviewing their distributions and inter-relationships with visualizations and making sure that the data meets the assumptions of the linear regression model. A model will be created using a wide variety of variables, which will then be reduced to just the explanatory variables which account for most of the changes to the dependent variable. 

### Four Assumptions of a Linear Regression Model

**<u>Assumption One:</u>** There is a linear relationship between the independent variable and the dependent variable. Linear relationships are constant and defined by the equation of a line: 

$$
y=mx + b
$$

Where y is the dependent variable, x is the independent variable, m is the slope or coefficient of change, and b is the value of the dependent variable when the independent variable is equal to zero. 

**<u>Assumption Two:</u>** There is low or no multicollinearity among independent variables. Multicollinearity occurs when two or more independent variables are highly correlated, meaning they essentially are explaining the same information. (Statology, 2021)

**<u>Assumption Three:</u>** Homoscedasticity or constant deviation of residuals. Errors or the difference between modeled and actual points should be constant, or show little deviation from zero. (Korstanje, 2020)

**<u>Assumption Four:</u>** Normality of residuals of model are normally distributed. Errors or the difference between modeled and actual points should be normally distributed, or symmetrically distributed near zero.

### Data Transformation 

The research question "What factors contribute to Additional Charges?" requires the analysis to look at any factors that might contribute positively or negatively to the variable 'Additional_charges'. To obtain that goal, we must consider data transformations that make the data easier to adapt to a linear regression model. First, categorical data of the boolean type with 'Yes' and 'No' answers will be encoded to numeric values 1 and 0. Second categorical variables both nominal and ordinal will be encoded with dummy variables so that they can be included in the model as independent variables. This will require first that these variables be converted into object or cateogorical type variables. 

```python
# Make a copy of the dataframe and examine the values in each column
mlmodeldf = reduceddf.copy()
mlmodeldf.head()
```

```python
# replace 'Yes'& 'No' values with 1 and 0 
mlmodeldf.HighBlood.replace(['Yes', 'No'], [1,0], inplace=True)
mlmodeldf.Overweight.replace(['Yes', 'No'], [1,0], inplace=True)
mlmodeldf.Diabetes.replace(['Yes', 'No'], [1,0], inplace=True)
mlmodeldf.Asthma.replace(['Yes', 'No'], [1,0], inplace=True)

# change variable type of categorical variables to object/category type
mlmodeldf.Initial_admin.astype('category')
mlmodeldf.Complication_risk.astype('category')
mlmodeldf.Services.astype('category')

# verify changes by viewing values in dataframe
mlmodeldf.head()
```

```python
# verify data type changes for categorical variables
mlmodeldf.dtypes
```

```python
# Use one-hot encoding technique to create dummy variables for each
# of the remaining categorical variables. 
mlmodeldf = pd.get_dummies(mlmodeldf, columns=["Complication_risk", "Services", 
                                               "Initial_admin"])
```

```python
# view dataframe with changes including dummy variables
mlmodeldf.head()
```

```python
# Drop one dummy variable column for each of the categories created above (Miles, 2021)
mlmodeldf = mlmodeldf.drop(columns = ['Complication_risk_Medium', 'Services_Intravenous', 
                           'Initial_admin_Elective Admission'])
mlmodeldf.head()
```

### Initial Linear Regression Model

The initial linear regression model aims to include a number of variables which could account for the changes in 'Additional_charges'. The model will use the .add_constant method included in the stats model library which will include a y-intercept into the model. 

```python
# Print column names for easy cut and paste 
mlmodeldf.columns
```

```python
#Initial Linear Regression Model 
import statsmodels.api as statsm
dependent = mlmodeldf['Additional_charges']
independent = mlmodeldf[['Age', 'Children', 'Income', 'Doc_visits',
       'Full_meals_eaten', 'vitD_supp', 'HighBlood', 'Overweight', 'Diabetes',
       'Asthma', 'Complication_risk_High', 'Complication_risk_Low',
       'Services_Blood Work', 'Services_CT Scan', 'Services_MRI',
       'Initial_admin_Emergency Admission',
       'Initial_admin_Observation Admission']].astype(float)

# Add a constant for y intercept 
independent = statsm.add_constant(independent)

# Run OLS linear regression 
model = statsm.OLS(dependent, independent).fit()

# View summary
model.summary()
```

### Model Reduction

### Justification of statistically based feature selection

The elimination of variables from the models will rely on statistical relationships. The first step will be to remove variables based on the multicollinearity assumption of linear regression. Independent variables with high multicollinearity will be removed. Next, a step-wise elimination based on statistical significance will be performed. First variables with p-values over .05 will be removed. Second, larger, but significant p-values will be removed and r-squared values compared for changes. Finally, variables will be removed individually, their r-squared values compared to find the model the best fit model that accounts for the highest amount of variation in 'Additional_charges'. 

### Reducing the model by checking for multicollinearity 

The first step in reducing the model comes from testing the assumption of multicollinearity among the independent variables. The goal of this step is to remove variables from the model that are strongly correlated to one another. 

```python
# VIF of Independent variables to check for multicollinearity (Geeks for Geeks, n.d.)
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Create dataframe of all independent variables
X = mlmodeldf[['Age', 'Children', 'Income', 'Doc_visits',
       'Full_meals_eaten', 'vitD_supp', 'HighBlood', 'Overweight', 'Diabetes',
       'Asthma', 'Complication_risk_High', 'Complication_risk_Low',
       'Services_Blood Work', 'Services_CT Scan', 'Services_MRI',
       'Initial_admin_Emergency Admission',
       'Initial_admin_Observation Admission']].astype(float)

# Run VIF for independent variables
{X.columns[i]: variance_inflation_factor(X.values, i) for i in range(1, X.shape[1])}
```

Independent variables with a value over 2 are removed from the model and the next iteration of the model is run without them. 

```python
# Run model without variables with high VIF (Over 2)
dependent = mlmodeldf['Additional_charges']
independent = mlmodeldf[['Age', 'Children',
       'Full_meals_eaten', 'vitD_supp', 'HighBlood', 'Diabetes',
       'Asthma', 'Complication_risk_High', 'Complication_risk_Low', 'Services_CT Scan',
        'Services_MRI','Initial_admin_Observation Admission']].astype(float)

independent = statsm.add_constant(independent)

model2 = statsm.OLS(dependent, independent).fit()
model2.summary()
```

### Results of first step - removing highly correlated independent variables
The second iteration, removing variables with high collinearity has neglibly changed the r-square factor from .0938 to .0937. 

### Step two - removing p-values over .05
The next step will to be to reduce the model by using the Wrapper Feature Selection method which aims to remove variables based on p-values. The next iteration will remove all variables with a p-value greater than .05

```python
# Run regression model with insignificant variables removed (p-value greater than .05)
dependent = mlmodeldf['Additional_charges']
independent = mlmodeldf[['Age', 'Children', 'HighBlood','Complication_risk_High', 
                         'Complication_risk_Low','Initial_admin_Observation Admission']].astype(float)

independent = statsm.add_constant(independent)

model3 = statsm.OLS(dependent, independent).fit()
model3.summary()
```

### Results from step 2
The R-squared for this model is identical to the second iteration. The variables can be removed without altering the outcomes. 

### Step 3 - Step-wise model reduction
Step-wise reduction by eliminating single variables at a time based on p-values. The first variable removed will be 'Children' since it has the highest p-value of remaining variables

```python
# remove children
dependent = mlmodeldf['Additional_charges']
independent = mlmodeldf[['Age','HighBlood','Complication_risk_High', 
                         'Complication_risk_Low','Initial_admin_Observation Admission']].astype(float)

independent = statsm.add_constant(independent)

model4 = statsm.OLS(dependent, independent).fit()
model4.summary()
```

Again there is no change in the r-square value which remains at .937. The next variable removed will be 'Complication_risk_Low' which is the dummy variable for Complication risk associated with 'Low' values.

```python
# remove children
dependent = mlmodeldf['Additional_charges']
independent = mlmodeldf[['Age','HighBlood','Complication_risk_High'
                         ,'Initial_admin_Observation Admission']].astype(float)

independent = statsm.add_constant(independent)

model5 = statsm.OLS(dependent, independent).fit()
model5.summary()
```

Again the r-square value is the same at .937 which means removing these variables has not  improved or reduced the model in any way. The next step will be to individually remove the remaining variables to see if removing any of them improves or reduces the model. 

```python
# Remove each remaining variable individually to compare r square 
dependent = mlmodeldf['Additional_charges']
independent_noage = mlmodeldf[['HighBlood','Complication_risk_High'
                         ,'Initial_admin_Observation Admission']].astype(float)
independent_noage = statsm.add_constant(independent_noage)
model6 = statsm.OLS(dependent, independent_noage).fit()
print("The R-Squared value for removing Age variable is: ", model6.rsquared)

independent_noHB = mlmodeldf[['Age', 'Complication_risk_High', 
                              'Initial_admin_Observation Admission']].astype(float)
independent_noHB = statsm.add_constant(independent_noHB)
model7 = statsm.OLS(dependent, independent_noHB).fit()
print("The R-Squared value for removing High Blood variable is: ", model7.rsquared)

independent_noCR = mlmodeldf[['Age', 'HighBlood','Initial_admin_Observation Admission']].astype(float)
independent_noCR = statsm.add_constant(independent_noCR)
model8 = statsm.OLS(dependent, independent_noCR).fit()
print("The R-Squared value for removing Complication Risk High variable is: ", model8.rsquared)

independent_noIA = mlmodeldf[['Age', 'HighBlood','Complication_risk_High']].astype(float)
independent_noIA = statsm.add_constant(independent_noIA)
model9 = statsm.OLS(dependent, independent_noIA).fit()
print("The R-Squared value for removing Initial Admin Observation variable is: ", model9.rsquared)
```

### Reduced Linear Regression Model

It is clear that the Age and High Blood pressure account for the greatest amount of variation in 'Additional_charges' and that removing either Complication Risk High or Initial Admin Observation Admission does not change the model in a significant way. Which brings us to the final reduced model. 

```python
dependent = mlmodeldf['Additional_charges']
independent = mlmodeldf[['Age','HighBlood']].astype(float)

independent = statsm.add_constant(independent)

model10 = statsm.OLS(dependent, independent).fit()
model10.summary()
```

### Comparison of Initial and Reduced Models

The initial model contains 17 variables, 13 plus dummy variables for categorical types. It has an R-squared value of .9379 which indicates these independent variables account for 93.8 percent of the variation in the dependent variable 'Additional_charges'. 

The final model contains two variables, 'HighBlood' and 'Age'. It has an R-Squared value of .9353 which indicates these independent variables account for 93.5 percent of the variation in the dependent variable 'Additional_charges'. 

The two models are nearly identical in their explanatory power for variation in the target variable. The extra variables in the initial model do not provide any additional explanation for the changes to 'Additional_charges'. Reasoning from the R-squared values, we can see that the other 15 variables accounted for very little variance in the dependent variable 'Additional_charges'. It is reasonable to say that the reduced model performs just as well as the initial model. In terms of simplicity the reduced model is superior, in that it reduces the the model without loss of explanation power. 


```python
# Show R-squared values for both models
print("R-squared value for initial model:", model.rsquared)
print("R-squared value for reduced model: ", model10.rsquared)
```

### Discusion of Results 

### Regression Equation 
The regression equation for the final model is: 

$$
 Y = -2677.6 + 225.7(Age) + 8637.3(HighBlood)
$$

### Interpretation of Coefficients

The co-efficients are adding dollar amounts to the prediction of 'Additional_costs'. The patient's age is multiplied by the coefficient for age of 225.70 dollars. Next, the 'HighBlood' coefficient is considered. Since this variable is boolean, 0 for no and 1 for yes, only patients with high blood pressure will have this coefficient added to their prediction. If the patient has high blood pressure the amount of 8637.30 dollars is added to the prediction. Finally, the y-intercept, identified in the model by the constant term is subtracted in the amount of 2677.60 dollars. 

For example, a patient who is 69 years of age and has high blood pressure is predicted to have (225.7)*69 + 8637.3 - 2677.6 = 5959.70 dollars in Additional Charges.


### Statistical and Practical Significance of Reduced Model

The statistical significance of the final model overall is okay, not great. R-Squared values are a good indication that the variance of 'Additional_charges' is mostly explained by the variables 'Age' and 'HighBlood'. It explains 93 percent of the variance overall. R-Squared values don't have defined limits for utility. There isn't a cutoff value that would indicate that this absolutely a useful model in terms of statistical significance. The residual calculations reveal the statistical limitations of the model. Overall, residuals conform mostly to a normal distribution but there is a problem with the predictions made at high and low Ages which cause the predicted values to be under-estimated. 

From a practical standpoint, the model is still useful. Both age and high blood pressure are both variables likely to be established at a patient's intake. From the very beginning of a patient's stay in the hospital it will be easy to establish, based on the coefficents of the model a likely amount to be accumulated throughout their stay. The simplicity of the model means that it is very easy to calculate a predicted total of additional costs. This has a number of practical benefits to stakeholders. From an administration perspective, knowing that patients of a higher age with blood-pressure are more likely to accumulate more charges, the hospital could undertake more targeted studies for these groups to understand the reasons for the charge and potentially identify possible cost-saving measures for both the hospital and patient. From the patient's perspective, this information is valuable as an incentive to manage high blood pressure later in life. Knowing that their high blood pressure could lead to greater costs in medical care could potentially be a motivation for self-management of the condition.

### Limitations of Data Analysis

The limitations have been discussed from a statistical standpoint previously, but it is important to note that the model does have problems with under-predictions for both patients with normal blood pressure at younger ages and patients with high blood pressure at older ages. Additionally these relationships between variables are simply correlated and do not indicate any type of causation.
